```python
import numpy as np
import matplotlib.pyplot as plt

N = 10_000

%matplotlib inline
```

```python
# x = np.random.randint(0, 10, 10_000)
x = np.linspace(0, 10, N + 1)
y = np.exp(-x / 10) * np.sin(x)

x, x.shape
```

```python
plt.plot(x, y)
```

```python
cond = (x >= 4) * (x <= 7)
res = y[cond]

np.mean(res), np.std(res)
```

```python
np.percentile(res, 80)
```

```python
plt.plot(
    np.gradient(y, x) # dy / dx
)
```

```python
plt.plot(
    x, np.gradient(y, x) # dy / dx
)
```

```python
dydx = np.gradient(y, x)
```

$$ P / (A \sigma \epsilon T_0^4) = \left( \frac{1}{1+e^{-kt}} \right)^4 $$

```python
kt = np.linspace(0, 3, 100)
P = (1 / (1 + np.exp(-kt))) ** 4
E = np.cumsum(P) * (kt[1] - kt[0])
```

```python
plt.plot(kt, P)
plt.xlabel(r'$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
```
